```python
df = pd.read_csv('boardgames1.csv')
```

```python
# 보드게임의 첫 출시는 UNO 로 1971년에 출시됐다
# 데이터에 1971년 이전에 발매됐다고 표시된게 2만개중 750개의 데이터이며 이는 무시할 수 없는 개수이므로 1971년 이전에 출시된 데이터에 결측값을 넣어준 후 평균을 구하고 평균값을 넣어줬습니다.

df.loc[df["yearpublished"]<1971 , "yearpublished"] = np.nan
df['yearpublished'].mean()
```

```python
df['yearpublished'].fillna(2005,inplace=True)
```

```python
# objectid 와 sortindex 의 성능이 겹치므로 objectid 열을 삭제
df.drop(['objectid'],axis=1,inplace=True)
# 단 하나의 unique 한 값을 가진 label 열을 삭제
df.drop(['label'],axis=1,inplace=True)
```

```python
# min_community 와 max_community 는 각각 커뮤니티 내에서 구한 평균 최소, 최대 플레이어이다. 그렇기 때문에 비인기 보드게임은 결측값이 존재한다.
# 그러므로 minplayers	maxplayers 는 publisher 가 정해준 최소, 최대 플레이어이므로 결측값을 채워줍니다.

df.loc[df['min_community'].isna(),['min_community']] = df['minplayers']
df.loc[df['max_community'].isna(),['max_community']] = df['maxplayers']
```

```python
# playerage 열 정제
# playerage 는 각각 커뮤니티 내에서 해당 보드게임을 플레이한 유저들의 최저 연령이다. 그러므로 결측값에 publisher가 정해준 값을 채워줍니다.
df['playerage'].unique()
```

```python
df['playerage'] = df['playerage'].str.replace('\\','', regex=True)
df['playerage'] = df['playerage'].str.replace(' and up','', regex=True)
df['playerage'] = df['playerage'].str.replace('Up to ','', regex=True)
tmp = df['playerage'].str.split('u',expand=True)
df['playerage'] = tmp[0]
```

```python
df.loc[df['playerage']=='(no votes)',['playerage']] = df['minage']
df.loc[df['playerage']=='All ages',['playerage']] = df['minage']
df['playerage'] = df['playerage'].astype(int)
```

```python
df['average'] = df['average'].astype(int) # 모델 적용을 위해 실수형 -> 정수형 변환
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, random_state=2)
```

```python
train, val = train_test_split(train, random_state=2)
```

```python
target = 'average'
```

```python
features = train.drop(columns=[target]).columns
```

```python
X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
X_test = test[features]
y_test = test[target]
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

pipe = make_pipeline(
    DecisionTreeClassifier(max_depth=3,random_state=1)
)

pipe.fit(X_train, y_train)
from sklearn.metrics import f1_score
predict = pipe.predict(X_val)
f1=f1_score(y_val, predict, average='micro')
print("F1 score: {}" .format(f1))
```
